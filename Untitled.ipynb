```python
import pandas as pd
import datetime
import yfinance as yf
from matplotlib import pyplot as plt
import mplcyberpunk
import smtplib
from email.message import EmailMessage
```

```python
#Pegar Dados do YFinance
```

```python
ativos = ["^BVSP", "BRL=X"]

hoje = datetime.datetime.now()

umAnoAtras = hoje - datetime.timedelta(days = 365)

dadosMercado = yf.download(ativos, umAnoAtras, hoje)

display(dadosMercado)
```

```python
#ManipulaÃ§Ã£o e ExclusÃ£o de Dados
```

```python
dadosFechamento = dadosMercado["Adj Close"]
dadosFechamento.columns=["Dolar", 'Ibovespa']
dadosFechamento = dadosFechamento.dropna()
dadosFechamento

```

```python
#Manipulando Dados - Criando tabelas com outros timeframes--nesse passo, foi isolado o dia final de cada mÃªs com seus respectivos fechamentos
```

```python
dadosFechamentoMensal = dadosFechamento.resample("M").last()
dadosFechamentoMensal
```

```python
dadosFechamentoAnual = dadosFechamento.resample("Y").last()
dadosFechamentoAnual
```

```python
#Calcular fechamento do dia, retorno anual e retorno diÃ¡rio dos Ativos
```

```python
retornoAnual = dadosFechamentoAnual.pct_change().dropna()
retornoMensal = dadosFechamentoMensal.pct_change().dropna()
retornoDiario = dadosFechamento.pct_change().dropna()
retornoDiario
```

```python
#Localizar o fechamento do dia anterior, retorno no mÃªs e no ano
#Loc= referencia e seleciona elementos a partir de seu nome
#Iloc= seleciona elementos a partir da lÃ³gica de uma matriz, com linhas e colunas
#EM PYTHON, A CONTAGEM SEMPRE COMEÃ‡A DO ZERO!!!
```

```python
retornoDiarioDolar = retornoDiario.iloc[-1,0]
retornoDiarioIbovespa= retornoDiario.iloc[-1,1]

retornoMensalDolar = retornoMensal.iloc[-1,0]
retornoMensalIbovespa= retornoMensal.iloc[-1,1]

retornoAnualDolar = retornoAnual.iloc[-1,0]
retornoAnualIbovespa= retornoAnual.iloc[-1,1]

retornoAnualIbovespa
```

```python
retornoDiarioDolar = round(retornoDiarioDolar*100, 2)
retornoDiarioIbovespa= round(retornoDiarioIbovespa*100, 2)

retornoMensalDolar = round(retornoMensalDolar*100, 2)
retornoMensalIbovespa= round(retornoMensalIbovespa*100, 2)

retornoAnualDolar = round(retornoAnualDolar*100, 2)
retornoAnualIbovespa= round(retornoAnualIbovespa*100, 2)

retornoAnualIbovespa


```

```python
  #Fazer Graficos da Performance dos Ativos
```

```python
plt.style.use("cyberpunk")
dadosFechamento.plot(y= 'Ibovespa', use_index= True, legend = False)
plt.title("Ibovespa")
mplcyberpunk.add_glow_effects()
plt.savefig('Ibovespa.png', dpi = 300)
plt.show()
```

```python
plt.style.use("cyberpunk")
dadosFechamento.plot(y= 'Dolar', use_index= True, legend = False)
plt.title("Dolar")
mplcyberpunk.add_glow_effects()
plt.savefig('Dolar.png', dpi = 300)
plt.show()
```
